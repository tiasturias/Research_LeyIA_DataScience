```coconut
from pathlib import Path
import pandas as pd
```

```coconut
# Import the analysis functions
from aei_analysis_functions_1p_api import (
    setup_plot_style,
    load_preprocessed_data,
    create_top_requests_bar_chart,
    create_platform_occupational_comparison,
    create_platform_lorenz_curves,
    create_collaboration_alluvial,
    create_automation_augmentation_panel,
    create_token_output_bar_chart,
    create_completion_vs_input_tokens_scatter,
    create_occupational_usage_cost_scatter,
    create_partial_regression_plot,
    perform_usage_share_regression_unweighted,
    create_btos_ai_adoption_chart,
)
```

```coconut
# Set matplotlib to use the correct backend and style
setup_plot_style()
```

```coconut
# Set up output directory for saving figures
output_dir = Path("../data/output/figures/")
btos_data_path = Path("../data/input/BTOS_National.xlsx")
api_data_path = Path("../data/intermediate/aei_raw_1p_api_2025-08-04_to_2025-08-11.csv")
cai_data_path = Path(
    "../data/intermediate/aei_raw_claude_ai_2025-08-04_to_2025-08-11.csv"
)

# Create output directory
output_dir.mkdir(parents=True, exist_ok=True)
```

```coconut
# Load BTOS Data
print("Loading BTOS data...")
btos_df = pd.read_excel(btos_data_path, sheet_name="Response Estimates")
btos_df_ref_dates_df = pd.read_excel(
    btos_data_path, sheet_name="Collection and Reference Dates"
)

# Load the API data
print("Loading API data...")
api_df = load_preprocessed_data(api_data_path)

# Load the Claude.ai data
print("Loading Claude.ai data...")
cai_df = load_preprocessed_data(cai_data_path)
```

```coconut
create_btos_ai_adoption_chart(btos_df, btos_df_ref_dates_df, output_dir)
```
